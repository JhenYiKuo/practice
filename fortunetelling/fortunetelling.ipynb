```python
#計算上中下格
LN = int(input('請輸入姓氏的康熙筆劃數'))
FN1 = int(input('請輸入名字第1個字的康熙筆劃數'))
FN2 = int(input('請輸入名字的第2個字的康熙筆劃數'))

up_pattern = int(1 + LN)
mid_pattern = int(LN + FN1)
down_pattern = int(FN1 + FN2)

#判斷屬性
up_five = up_pattern % 10
if up_five == 1 or up_five == 2:
    up_element = '木'
elif up_five == 3 or up_five == 4:
    up_element = '火'
elif up_five == 5 or up_five == 6:
    up_element = '土'
elif up_five == 7 or up_five == 8:
    up_element = '金'
else:
    up_element = '水'

mid_five = mid_pattern % 10
if mid_five == 1 or mid_five == 2:
    mid_element = '木'
elif mid_five == 3 or mid_five == 4:
    mid_element = '火'
elif mid_five == 5 or mid_five == 6:
    mid_element = '土'
elif mid_five == 7 or mid_five == 8:
    mid_element = '金'
else:
    mid_element = '水'

down_five = down_pattern % 10
if down_five == 1 or down_five == 2:
    down_element = '木'
elif down_five == 3 or down_five == 4:
    down_element = '火'
elif down_five == 5 or down_five == 6:
    down_element = '土'
elif down_five == 7 or down_five == 8:
    down_element = '金'
else:
    down_element = '水'

print(f'你的上格是{up_pattern}，五行屬{up_element}')
print(f'你的中格是{mid_pattern}，五行屬{mid_element}')
print(f'你的下格是{down_pattern}，五行屬{down_element}')

#判斷思想功能
if (up_element, mid_element) in [('木','木'),('木','火'),('火','木'),('金','木'),('木','金')]:
    think_mode ='木'
elif (up_element, mid_element) in [('火','火'),('火','土'),('土','火'),('水','火'),('火','水')]:
    think_mode ='火'
elif (up_element, mid_element) in [('土','土'),('土','金'),('金','土'),('木','土'),('土','木')]:
    think_mode ='土'
elif (up_element, mid_element) in [('金','金'),('金','水'),('水','金'),('火','金'),('金','火')]:
    think_mode ='金'
elif (up_element, mid_element) in [('水','水'),('水','木'),('木','水'),('土','水'),('水','土')]:
    think_mode ='水'

#判斷思想快慢
if think_mode in ['木','火']:
    think_way = '快'
elif think_mode in ['土','金','水']:
    think_way = '慢'

print(f'你的思想功能屬於{think_mode}型人，思考得{think_way}')

#判斷行動功能
if (mid_element, down_element) in [('木','木'),('木','火'),('火','木'),('金','木'),('木','金')]:
    act_mode ='木'
elif (mid_element, down_element) in [('火','火'),('火','土'),('土','火'),('水','火'),('火','水')]:
    act_mode ='火'
elif (mid_element, down_element) in [('土','土'),('土','金'),('金','土'),('木','土'),('土','木')]:
    act_mode ='土'
elif (mid_element, down_element) in [('金','金'),('金','水'),('水','金'),('火','金'),('金','火')]:
    act_mode ='金'
elif (mid_element, down_element) in [('水','水'),('水','木'),('木','水'),('土','水'),('水','土')]:
    act_mode ='水'

#判斷行動快慢
if act_mode in ['木','火']:
    act_way = '快'
elif act_mode in ['土','金','水']:
    act_way = '慢'

print(f'你的行動功能屬於{act_mode}型人，行動得{act_way}')
```
