```python
#計算八卦氣
def get_eightair(end, start):
    if end < start: #進位以確保相減>0
        end += 9
    if end >= start:
        if end - start == 0:
            return '中'
        elif end - start == 1:
            return '乾'
        elif end - start == 2:
            return '兌'
        elif end - start == 3:
            return '艮'
        elif end - start == 4:
            return '離'
        elif end - start == 5:
            return '坎'
        elif end - start == 6:
            return '坤'
        elif end - start == 7:
            return '震'
        elif end - start == 8:
            return  '巽'
        
#取化氣數
def digit_sum(number):
    while number >= 10:
        number = sum(int(digit) for digit in str(number))
    return number
```

```python
#八卦氣

from datetime import datetime
current_datetime = datetime.now()
current_year = current_datetime.year - 1911 #民國年

#取得上下中總的化氣數(數字為頭)
up_digit = digit_sum(up_pattern)
mid_digit = digit_sum(mid_pattern)
down_digit = digit_sum(down_pattern)
all_digit = digit_sum(all_pattern)

#取得當年民國年的化氣數(日子為尾)
year_digit = digit_sum(current_year)

#計算八卦氣
up_eightair = get_eightair(year_digit, up_digit)
mid_eightair = get_eightair(year_digit, mid_digit)
down_eightair = get_eightair(year_digit, down_digit)
all_eightair = get_eightair(year_digit, all_digit)

print(f'您今年的上格八卦氣碰{up_eightair}')
print(f'您今年的中格八卦氣碰{mid_eightair}')
print(f'您今年的下格八卦氣碰{down_eightair}')
print(f'您今年的總格八卦氣碰{all_eightair}')
```
